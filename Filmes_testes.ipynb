# Budget GERAL

```python
tfB = tf[tf.budget>0]

tfB['release_date']=tfB['release_date'].str[6:10]
a = pd.to_numeric(tfB.release_date,errors='coerce')
tfB.release_date = a
tfB = tfB[tfB.release_date>=1960]

index_year1 = tfB["release_date"].map(lambda x:x)
df_b = tfB.groupby(pd.Index(index_year1)).sum()[['budget']]


df_b.head()
```
